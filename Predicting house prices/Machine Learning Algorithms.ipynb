# VERİ ÖN İŞLEME

```python
y= df_train.SalePrice
x= df_train.drop(["SalePrice"],axis=1)
```

y bizim hedef niteliğimiz olduğu için ayrı bir değişkene ve x ise hedef niteliğimizi etkileyen öz nitelikler olduğu için farklı bir değişkene atıldı. Girdi ve çıktı değişkenlerini ayarladık. Girdi ve çıktı değişkenleri train ve test olarak ayrılır. Train ile model kurulur test ile model değerlendirilir. Validasyon değerleri ile de hiper parametreler ayarlanır.
```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_val, y_train, y_val= train_test_split(x,y,train_size=0.8,random_state=0 )
```

Bir makine öğrenmesi algoritmasını analiz etmeden önce sütunlarını hazır hale getirmek gerekiyor.

```python
categorical_cols= [cname for cname in x_train.columns if x_train[cname].nunique()<10 and x_train[cname].dtype=="object"  ]
```

Yukarıda kategorik verileri filtreleme işlemi uyguladık, nasıl uyguladık peki öncelikle veritipimizin object olanlar ve kategori adımınız çeşidinin 10'dan küçük olan kategorik sütunları seçtik ve bir değişkene atadık.

```python
len(categorical_cols)
```

```python
numerical_cols= [cname for cname in x_train.columns if  x_train[cname].dtype in ["int64", "float64"]]
```

```python
len(numerical_cols)
```

Totalde 70 sütun seçtik demek ki seçilmeyen diğer sutünlar kategorik değeri 10 dan fazla. Bunun için o sütunları çıkaralım.

```python
my_cols= categorical_cols + numerical_cols
```

```python
x_train= x_train[my_cols]
x_val=x_val[my_cols]
x_test=df_test[my_cols]
```

# PİPELİNE KURMA

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

Pipeline nedir, pipeline yapılacak işlemleri bir senkron şeklinde ilerletmek için kurduğumuz bir yapıdır. Belirledğimiz işlem ve o işlem yapılırken kullanılacak stratejiyi belirledik.

numerical transformers= imputer num adlı bir nesne oluşturduk ve boş verilerin yerine (numeric verilerin yerine) o sütunun median değerinin koyulmasını ve scaler nesnesi oluşturup sayısal değerlerde ölçekleme yapması için Standart Scaler kullandık.

```python
numerical_transformer = Pipeline(steps=[
    ("imputer_num", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
```

Kategorik veriler için eksik veriler yerine mod değeri yani en çok tekrar eden değer yazılır. Kategorik verileri sayısal değerlere çevirmek için OneHotEncoder kullandık. Bilinmeyen bir kategori ile karşılaştığında handle_unknown değerini ignore ile eşitledik.

```python
categorical_transformer = Pipeline(steps = [
    ("imputer_cal", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])
```

Veri ön işlemenin neredeyse sonuna geldik hem sayılsal veriler için hem de kategorical veriler için kurduğumuz pipelineları öncesinde belirttiğimiz numerical_cols ve categorical_cols adlı değişkenler üzerinde uygulamaya geçelim.  

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer(transformers=[
    ("num", numerical_transformer, numerical_cols),
    ("cat", categorical_transformer, categorical_cols)
])
```

# MODEL KURMA

# Random Forest Algoritması

Random Forrest Ensemble bir algoritmadır.

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state=0)
```

```python
#100 ağaçlık bir random forest algoritması oluşturalım.

rf= RandomForestRegressor(n_estimators=100, random_state=0)
```

```python
my_pipeline = Pipeline(steps=[ ("preprocessor", preprocessor),("model", rf)])
```

Modeli eğitmek için fit methodunun kullanıyoruz.

```python
my_pipeline.fit(x_train, y_train)
```

# MODEL OLUŞTURMA

Validasyon verilerini tahmin edelim.

```python
val_preds = my_pipeline.predict(x_val)
val_preds
```

Hedef değişkenimiz sayısal verilerdir. Bu sebeple problemimiz Regresyon problemi haline gelir. Regresyon probleminin en önemli metriklerden biri mead absolute error metriğidir. Modelimizin validasyon verisi üzerinde performansını görelim.

```python
from sklearn.metrics import mean_absolute_error
print("Validation MAE: ", mean_absolute_error(y_val, val_preds)) #y val değerler, val_pred tahmin ettiğimz
```
